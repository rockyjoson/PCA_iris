```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
from sklearn.metrics import classification_report
df=pd.read_csv('/content/drive/MyDrive/DataSets_ML/IRIS.csv')
df
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['species']=le.fit_transform(df['species'])
df
```

```python
X=df.iloc[:,:-1]
X
```

```python
y=df.iloc[:,-1]
y
```

```python
mm=MinMaxScaler()
X=mm.fit_transform(X)
X
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=1)
X_train
```

```python
pca=PCA(n_components=3,random_state=1) # default no: of components= no of features=4
X_train=pca.fit_transform(X_train) # eigen vectors generated using X_train
X_test=pca.transform(X_test)
X_train   # 3 components generated
  # this step can be done before splitting also with X only
```

```python
pca.explained_variance_ratio_  # to analyze the amount of information in each principal component # highest PC1
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
knn.fit(X_train,y_train)
y_pred=knn.predict(X_test)
print(classification_report(y_test,y_pred))
```
